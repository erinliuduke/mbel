```python
import json
import openai
import time
```

```python
f = open('train.json')
d = json.load(f)
f.close()

i = 30
da = d[i]['story'].split(". ")
```

```python
openai.api_key = "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"
system_msg = input("What type of chatbot would you like to create?\n")
print("Your new assistant is ready!")

while input != "quit()":
    captdict = []
    for j in range(50):
        k = j+6
        da = d[k]['story'].split(". ")
        messages = []
        capts = []
        
        for i in range(len(da)):
            
            messages.append({"role": "system", "content": system_msg})
            message = f"make '{da[i]}' a very concise image description that must include a consistent detailed animal description about the animal protagonist"
            print(message)
            
            messages.append({"role": "user", "content": message})
            
            response = openai.ChatCompletion.create(
                model="gpt-3.5-turbo",
                messages=messages, temperature = 0)
            reply = response["choices"][0]["message"]["content"]
            messages.append({"role": "assistant", "content": reply})
            print("\n" + reply + "\n")
            
            capts.append({"story segment": da[i], "caption": reply})
            time.sleep(21)
            
        captdict.append(capts)
        with open(f'storyseg-caption{k}.json', 'w') as fp:
            json.dump(capts, fp)
        
```
